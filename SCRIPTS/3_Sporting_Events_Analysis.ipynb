```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.dates as mdates

#This code created with the help of ChatGPT
```

```python
#import csv and change Date column to Date format
DKNG = pd.read_csv("/content/dkng_ts.csv",parse_dates=['Date'], index_col='Date')
```

```python
DKNG.head()
# check and make sure data loads correctly
```

Create plain time series

```python
sns.set(style="darkgrid")  # Set Seaborn style
plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='DKNG.Adjusted', data=DKNG) #create basic time series

#since we have so many dates, need to change the x axis to only show every three months to prevent overcrowding
ax = plt.gca()
ax.xaxis.set_major_locator(mdates.MonthLocator(interval=3))  # Tick every 3 months
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))  # Format as 'Jan 2023' for readability

# Customize the plot and add labels
plt.title('Time Series of DKNG adjusted stock')
plt.xlabel('Date')
plt.ylabel('Price')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

# Show the plot
plt.show()
```

Add shaded regions that correspond to major sporting events - checking to see if peaks have any relation to these events


```python
sns.set(style="darkgrid")  # Set Seaborn style
plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='DKNG.Adjusted', data=DKNG)

ax = plt.gca()
ax.xaxis.set_major_locator(mdates.MonthLocator(interval=3))  # Tick every 3 months
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))  # Format as 'Jan 2023'


# Add shaded regions - these dates were found by looking up major dates in the sporting world

#superbowl
plt.axvspan(pd.Timestamp('2024-02-09'), pd.Timestamp('2024-02-15'), color='red', alpha=0.2, label='Super Bowl')

#world series
plt.axvspan(pd.Timestamp('2024-10-25'), pd.Timestamp('2024-11-02'), color='blue', alpha=0.2, label='World Series')

#nba finals + stanley cup
plt.axvspan(pd.Timestamp('2024-06-05'), pd.Timestamp('2024-06-24'), color='green', alpha=0.2, label='NBA Finals / Stanley Cup')

#march madness
plt.axvspan(pd.Timestamp('2024-03-19'), pd.Timestamp('2024-04-08'), color='yellow', alpha=0.2, label='March Madness')



# Customize the plot
plt.title('Time Series of DKNG adjusted stock')
plt.xlabel('Date')
plt.ylabel('Price')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

# Legend
plt.legend()

# Show the plot
plt.show()
```



*   Seeing big dip in 2022 - this corresponds to when california voted down a bill that would've expanded online sports betting (see online resources)
*   Biggest peak matches with march madness in 2024 - does this peak remain the same with other years?
* Small local peaks during every sporting event but the world series
